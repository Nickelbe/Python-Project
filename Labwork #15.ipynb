```python
import matplotlib.pyplot as plt
import numpy as np
from math import *
import matplotlib.ticker

fig , T = plt.subplots()

plt.title('Лабораторная работа №15')
plt.ylabel('угол , а\u00b0')
plt.xlabel('угол , в\u00b0')

x = []
y = []
x1 = []
y1 = []

xerr = 2
yerr = 2

f = open('Data1.txt')
c = 0
for line in f:
        nums = [float(i) for i in line.split()]
x1.append(nums[2])
        y1.append(nums[3])
        c = c+1
        if c == 20:
            break
        x.append(nums[0])
        y.append(nums[1])
        nums = [float(i) for i in line.split()]
x1.append(nums[2])
        y1.append(nums[3])
        c = c+1
        if c == 20:
            break
T1 = T.plot(x,y,color = 'red',linewidth = 1)
T.errorbar(x, y, xerr=xerr, yerr=yerr, fmt='--', ecolor='black')
T2 = T.plot(x1,y1,color = 'blue',linewidth = 1)
T.errorbar(x1, y1, xerr=xerr, yerr=yerr, fmt='--', ecolor='black')
T.grid()

plt.ylim(-10, 120)
plt.xlim(0,180)
plt.tight_layout()
plt.show()
```
