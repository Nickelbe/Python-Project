```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker
import random

fig , T = plt.subplots()

plt.title('Лабораторная работа №15')
plt.ylabel('угол , в\u00b0')
plt.xlabel('угол , а\u00b0')

x = []
y = []

xerr = 2
yerr = 2

f = open('Data1.txt')
c = 0
for line in f:
    nums = [float(i) for i in line.split()]
    x.append(nums[0])
    y.append(nums[1])
    c = c+1
    if c == 20:
        break
T.plot(x,y ,color = 'red')
T.errorbar(x, y, xerr=xerr, yerr=yerr, ecolor = 'black')
T.grid()
plt.tight_layout()
plt.show()
```
